```python
import numpy as np
import pandas as pd
import csv
import os
```

```python
folder = "../../Data/Precinct_Nov_6/"
values = []
for file in os.listdir(folder):
    fin = folder + "/" + str(file)
    print(fin)
    
    xl = pd.ExcelFile(fin)
    df = xl.parse('DEMOGRAPHICS REPORT')
    for row in range(2, df.shape[0]):
        values_county = [
                  file.strip(".xlsx"), # County
                  str(df.iloc[row,0]).lower().lstrip("0"), # Precinct ID 
                  str(df.iloc[row,1]).lower(), # Precinct name 
                  df.iloc[row,2], # Black Male registered
                  df.iloc[row,3], # Black Male voted                  
                  df.iloc[row,5], # Black Female registered
                  df.iloc[row,6], # Black Female voted                      
                  df.iloc[row,8], # Black Unknown registered
                  df.iloc[row,9], # Black Unknown voted                      
                  df.iloc[row,11],# White Male registered
                  df.iloc[row,12],# White Male voted                  
                  df.iloc[row,14],# White Female registered
                  df.iloc[row,15],# White Female voted  
                  df.iloc[row,17],# White Unknown registered
                  df.iloc[row,18],# White Unknown voted
                  df.iloc[row,20],# Asian Male registered
                  df.iloc[row,21],# Asian Male voted                  
                  df.iloc[row,23],# Asian Female registered
                  df.iloc[row,24],# Asian Female voted  
                  df.iloc[row,26],# Asian Unknown registered
                  df.iloc[row,27],# Asian Unknown voted 
                  df.iloc[row,29],# Hispanic Male registered
                  df.iloc[row,30],# Hispanic Male voted                  
                  df.iloc[row,32],# Hispanic Female registered
                  df.iloc[row,33],# Hispanic Female voted  
                  df.iloc[row,35],# Hispanic Unknown registered
                  df.iloc[row,36],# Hispanic Unknown voted 
                  df.iloc[row,38],# Native American Male registered
                  df.iloc[row,39],# Native American Male voted                  
                  df.iloc[row,41],# Native American Female registered
                  df.iloc[row,42],# Native American Female voted  
                  df.iloc[row,44],# Native American Unknown registered
                  df.iloc[row,45],# Native American Unknown voted 
                  df.iloc[row,47],# Other Male registered
                  df.iloc[row,48],# Other Male voted                  
                  df.iloc[row,50],# Other Female registered
                  df.iloc[row,51],# Other Female voted  
                  df.iloc[row,53],# Other Unknown registered
                  df.iloc[row,54],# Other Unknown voted 
                  df.iloc[row,56],# Unknown Male registered
                  df.iloc[row,57],# Unknown Male voted                  
                  df.iloc[row,59],# Unknown Female registered
                  df.iloc[row,60],# Unknown Female voted  
                  df.iloc[row,62],# Unknown registered
                  df.iloc[row,63],# Unknown voted 
                  df.iloc[row,65],# Total registered
                  df.iloc[row,66] # Total voted        
                 ]
        values.append(values_county)
        print(values_county)

```

```python
with open('../../Data/turnout_by_precinct.csv', mode='w') as turnout_file:
    csv_writer = csv.writer(turnout_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    csv_writer.writerow(["County", "Precinct_ID", "Precinct_name",
                         "Black_Male_reg", "Black_Male_voted",
                         "Black_Female_reg", "Black_Female_voted",
                         "Black_unknown_reg", "Black_unknown_voted",
                         "White_Male_reg", "White_Male_voted",
                         "White_Female_reg", "White_Female_voted",
                         "White_unknown_reg", "White_unknown_voted",
                         "Asian_Male_reg", "Asian_Male_voted",
                         "Asian_Female_reg", "Asian_Female_voted",
                         "Asian_unknown_reg", "Asian_unknown_voted",
                         "Hisp_Male_reg", "Hisp_Male_voted",
                         "Hisp_Female_reg", "Hisp_Female_voted",
                         "Hisp_unknown_reg", "Hisp_unknown_voted",
                         "NAm_Male_reg", "NAm_Male_voted",
                         "NAm_Female_reg", "NAm_Female_voted",
                         "NAm_unknown_reg", "NAm_unknown_voted",
                         "Other_Male_reg", "Other_Male_voted",
                         "Other_Female_reg", "Other_Female_voted",
                         "Other_unknown_reg", "Other_unknown_voted",
                         "Unknown_Male_reg", "Unknown_Male_voted",
                         "Unknown_Female_reg", "Unknown_Female_voted",
                         "Unknown_unknown_reg", "Unknown_unknown_voted",
                         "Total_reg", "Total_voted"])
    for item in values:
        csv_writer.writerow(item)

turnout_file.close()
```
